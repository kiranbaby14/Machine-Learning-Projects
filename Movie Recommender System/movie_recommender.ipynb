```python
#import the required libraries
import numpy as np
import pandas as pd
```

```python
#read the csv to dataframes
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
#movies dataframe
movies.head(1)
```

```python
#credits dataframe
credits.head(1)
```

```python
movies.shape
```

```python
#merge movies and credits dataframe into one
movies = movies.merge(credits, on='title')
movies.shape
```

```python
#select only necesary columns
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head()
```

```python
movies.info()
```

```python
#get the rows which are empty
movies.isnull().sum()
```

```python
movies.shape
```

```python
#Drop the rows which have null values
movies.dropna(inplace=True)
movies.shape
```

```python
#check for duplicate rows
movies.duplicated().sum()
```

```python
movies.iloc[0]
```

```python
import ast #converts string of list --> list
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
convert('[{"id": 12, "name":"anjwbyu"}]')
```

```python
# apply convert() to movies dataframe --> 'genres column only'
movies['genres'] = movies['genres'].apply(convert)
```

```python
# apply convert() to movies dataframe --> 'keywords column only'
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
import ast #converts string of list --> list
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            L.append(i["name"])
        count += 1
    return L
```

```python
# apply convert() to movies dataframe --> 'cast column only && return the top 3 casts'
movies['cast'] = movies['cast'].apply(convert3) 
movies['cast'].head(3)
```

```python
#function to return the director of the movie
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])

movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#import vectorization class from sklearn and put max 5000 words and remove joining words in english
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0].shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
